```python
data1.columns=['Patient ID #', 'Sample ID #', 'Tumor type', 'AJCC Stage',
       'AFP', 'Angiopoietin-2', 'AXL', 'CA125',
       'CA15-3', 'CA19-9', 'CD44', 'CEA',
       'CYFRA 21-1', 'DKK1', 'Endoglin',
       'FGF2', 'Follistatin', 'Galectin-3',
       'G-CS', 'GDF15', 'HE4', 'HGF',
       'IL-6', 'IL-8', 'Kallikrein-6',
       'Leptin', 'Mesothelin', 'Midkine ',
       'Myeloperoxidase', 'NSE', 'OPG', 'OPN',
       'PAR', 'Prolactin', 'sEGFR', 'sFas',
       'SHBG', 'sHER2/sEGFR2/sErbB2', 'sPECAM-1',
       'TGFa', 'Thrombospondin-2', 'TIMP-1',
       'TIMP-2','CancerSEEK Logistic Regression Score',
       'CancerSEEK Test Result', 'REGRESSION_SCORE']
```

```python
data1
```

```python
dict_vals = {'AFP': 20000.0,
 'CA125': 35.0,
 'CA15-3': 31.3,
 'CA19-9': 35,
 'CEA': 5200.0,
 'Prolactin': 23300.0,
 }
```

```python
data1.shape
```

```python
data1.loc[:,'REGRESSION_SCORE'] = np.zeros(1817)
```

```python
data1.dtypes
```

```python
dict_vals.keys()
```

```python
def removes(x):
  x=x.replace('*','')
  x=x.replace(' ','')
  return x

for i in dict_vals.keys():
  data1[i]=data1[i].apply(removes)
```

```python
data1
```

```python
data1.columns
```

```python
dict_vals
```

```python
def find_protein(x):
  max_count=0
  for i in dict_vals.keys():
    if float(x[i])>=dict_vals[i]:
      max_count+=1
  if max_count>=1:
    return 1
  else:
    return 0
```

```python
data1['REGRESSION_SCORE']=data1.apply(lambda row:find_protein(row),axis=1)
```

```python
from collections import Counter

a=Counter(list(data1['REGRESSION_SCORE']))
a
```

```python
data1.head()
```

```python
x=data1[['AFP', 'CA125', 'CA15-3', 'CA19-9', 'CEA', 'Prolactin']]
y=data1['REGRESSION_SCORE']
```

```python
x
```

```python
x.dtypes
```

```python
def convert_type(x):
  x=float(x)
  return x

for i in x.columns:
  x[i]=x[i].apply(convert_type)
```

```python
x.dtypes
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import confusion_matrix, accuracy_score 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=15)

logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)*100

confusion_mat = confusion_matrix(y_test,y_pred)

print("Accuracy is",accuracy)
print("Confusion Matrix")
print(confusion_mat)
```

```python
y_predict_prob = logistic_model.predict_proba(x_test)
```

```python
y_predict_prob_class_1 = y_predict_prob[:,1]
```

```python
y_predict_prob_class_1
```

```python
y_predict_class = [1 if prob > 0.892 else 0 for prob in y_predict_prob_class_1]
```

```python
print("Accuracy:", round(accuracy_score(y_test, y_predict_class), 3))
```
